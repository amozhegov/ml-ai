```python
# for Jupyter-book, we copy data from GitHub, locally, to save Internet traffic,
# you can specify the data/ folder from the root of your cloned
# https://github.com/Yorko/mlcourse.ai repo, to save Internet traffic
DATA_URL = "https://raw.githubusercontent.com/Yorko/mlcourse.ai/main/data/"
```

```python
ds = pd.read_csv(DATA_URL + "adult.data.csv")
ds.head()
```

1. How many men and women (sex feature) are represented in this dataset?

```python
print('Amount of males in the Dataset are:', ds['sex'].to_list().count('Male'))
```

```python
print('Amount of females in the Dataset are:', ds['sex'].to_list().count('Female'))
```

2. What is the average age (age feature) of women?

```python
ds.loc[ds['sex'] == 'Female', 'age'].mean()
```

3. What is the percentage of German citizens (native-country feature)?

```python
ds.loc[ds['native-country'] == 'Germany']
```

```python
ds['native-country'].to_list().count('Germany')
```
