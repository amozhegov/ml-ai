4. What are the mean and standard deviation of age for those who earn more than 50K per year (salary feature) and those who earn less than 50K per year?

```python
mean = statistics.mean(ds['salary'] == '>50K')
stand_dev = statistics.stdev(ds['salary'] == '>50K')

print('Mean and Standard deviation of age for those who earn more than 50K per year:')
print('Mean:', mean)
print('Standard deviation:', stand_dev)

mean = statistics.mean(ds['salary'] == '<=50K')
stand_dev = statistics.stdev(ds['salary'] == '<=50K')
print()
print('Mean and Standard deviation of age for those who earn less than 50K per year:')
print('Mean:', mean)
print('Standard deviation:', stand_dev)
```
