```python
# for Jupyter-book, we copy data from GitHub, locally, to save Internet traffic,
# you can specify the data/ folder from the root of your cloned
# https://github.com/Yorko/mlcourse.ai repo, to save Internet traffic
DATA_URL = "https://raw.githubusercontent.com/Yorko/mlcourse.ai/main/data/"
```

```python
ds = pd.read_csv(DATA_URL + "adult.data.csv")
ds.head()
```

1. How many men and women (sex feature) are represented in this dataset?

```python
print('Amount of males in the Dataset are:', ds['sex'].to_list().count('Male'))
```

```python
print('Amount of females in the Dataset are:', ds['sex'].to_list().count('Female'))
```

2. What is the average age (age feature) of women?

```python
ds.loc[ds['sex'] == 'Female', 'age'].mean()
```

3. What is the percentage of German citizens (native-country feature)?

```python
ds.loc[ds['native-country'] == 'Germany']
```

```python
ds_germans_share = ds['native-country'].to_list().count('Germany') / len(ds) * 100
precision = 2
print('Percentage of German citizens in the Dataset is ', format(ds_germans_share, f'.{precision}f'), '%', sep='')
```

4. What are the mean and standard deviation of age for those who earn more than 50K per year (salary feature) and those who earn less than 50K per year?

```python
mean = statistics.mean(ds['salary'] == '>50K')
stand_dev = statistics.stdev(ds['salary'] == '>50K')

print('Mean and Standard deviation of age for those who earn more than 50K per year:')
print('Mean:', mean)
print('Standard deviation:', stand_dev)

mean = statistics.mean(ds['salary'] == '<=50K')
stand_dev = statistics.stdev(ds['salary'] == '<=50K')
print()
print('Mean and Standard deviation of age for those who earn less than 50K per year:')
print('Mean:', mean)
print('Standard deviation:', stand_dev)
```
